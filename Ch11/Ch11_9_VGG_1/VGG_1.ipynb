```python
#VGG16の取得
#VGG16は、高い精度で画像を分類できるモデルとして有名
#VGG16の取得
#まず、KerasのライブラリからVGG16をインポートし、変数に代入します。引数にweights='imagenet'と指定することで、
#ImageNetデータセットで事前学習されたパラメータ（重みと閾値）を持つ予測モデルを読み込めます。
from tensorflow.keras.applications.vgg16 import VGG16
model = VGG16(weights='imagenet')

#VGG16の可視化
#次に、model.summary()でVGG16の予測モデルの構造を確認
model.summary()

#load_img関数で画像を取り込みます。target_sizeパラメータで、取り込み時の画像のサイズを指定
#target_sizeとして縦224ピクセル×横224ピクセルを指定
from tensorflow.keras.preprocessing.image import load_img
unknown_img = load_img('sample_data/unknown_vgg.jpg', target_size=(224, 224))
unknown_img

#サンプルデータの前処理
#続いて、画像データを予測モデルに入力する前に、適切な形式へ変換
from tensorflow.keras.preprocessing.image import img_to_array
unknown_array = img_to_array(unknown_img)
unknown_array.shape

#インプットデータの構造は「任意のサンプル数×224×224×3」です。reshape関数で同じデータ構造に変換
unknown_array = unknown_array.reshape((1, 224, 224, 3))
unknown_array.shape

#VGG16のpreprocess_input関数を利用して、前処理を実施
from tensorflow.keras.applications.vgg16 import preprocess_input
unknown_array = preprocess_input(unknown_array)

#予測
#次に、VGG16を使って画像データの予測
result = model.predict(unknown_array)
result

#decode_predictions関数を利用すると、可能性の高い上位5つのカテゴリと、そのカテゴリに属する確率が表示され
from tensorflow.keras.applications.vgg16 import decode_predictions
decode_predictions(result)
```
